### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
chipo.head(10)
```

### Step 4. How many products cost more than $10.00?

```python
chipo['item_price_converted'] = chipo['item_price'].apply(lambda x: float(x[1:]))
# item_price = chipo.groupby('item_name')['item_price'].max()
# chipo.info()
item_price_pair = chipo.groupby(['item_name'])['item_price_converted'].max().reset_index()
item_price_pair.head(3)
# print(item_price_pair.shape[0])
len(item_price_pair[item_price_pair['item_price_converted'] > 10])
# item_price_pair[ item_price_pair['item_price_converted'] > 10 ].shape[0]

# Solution:
price = [float(x[1:]) for x in chipo.item_price]
chipo['price'] = price

chipo_dedup = chipo.drop_duplicates(['item_name', 'quantity'])
chipo_filtered = chipo_dedup[ chipo_dedup.quantity == 1]
chipo_filtered[ chipo_filtered.price > 10].item_name.nunique()
```
