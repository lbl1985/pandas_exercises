### Step 4. For the purpose of this exercise slice the dataframe from 'school' until the 'guardian' column

```python
column_name = list(df.columns)
print(column_name.index('school'))
print(column_name.index('guardian'))

df = df.iloc[:, range(column_name.index('school'), column_name.index('guardian') + 1)]
df.head(3)
```

### Step 5. Create a lambda function that capitalize strings.

```python
df.Mjob = df.Mjob.apply(lambda x: x.upper())
df.loc[:, ['Mjob', 'Fjob', 'reason', 'guardian']] = df.loc[:, ['Mjob', 'Fjob', 'reason', 'guardian']].apply(lambda x: x.apply(lambda y: y.upper()))
df.head(3)
```

### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
def majority(x):
    return True if x > 17 else False

df['legal_drinker'] = df.age.apply(majority)
df.head(3)
```
